#Static Processing

```
from pyspark.sql.types import *

#input path with 1000 json
inputPath = "/FileStore/tables/sfd/"

# Explicitly set schema
schema = StructType([ StructField("address", StringType(), True),
                      StructField("type", StringType(), True),
                      StructField("datetime", StringType(), True),
                      StructField("latitude", StringType(), True),
                      StructField("longitude", StringType(), True),
                      StructField("report_location", StringType(), True),
                      StructField("incident_number", StringType(), True),
                      StructField(":@computed_region_ru88_fbhk", StringType(), True),
                      StructField(":@computed_region_kuhn_3gp2", StringType(), True),
                      StructField(":@computed_region_q256_3sug", StringType(), True),
                    ])

# Static DataFrame representing data in the JSON files
staticInputDF = (
  spark
    .read
    .schema(schema)
    .json(inputPath)
)

display(staticInputDF)
```

##Data wrangling

```
df = staticInputDF
```

```
import pyspark.sql.functions as fc
from pyspark.sql.types import IntegerType


df = df.withColumn(":@computed_region_ru88_fbhk", df[":@computed_region_ru88_fbhk"].cast(IntegerType()))
df = df.withColumn(":@computed_region_kuhn_3gp2", df[":@computed_region_kuhn_3gp2"].cast(IntegerType()))
df = df.withColumn(":@computed_region_q256_3sug", df[":@computed_region_q256_3sug"].cast(IntegerType()))
df = df.withColumn("latitude", col("latitude").cast("float"))
df = df.withColumn("longitude", col("longitude").cast("float"))
```

```
#analisi dei null e dei nan

df.agg(*[fc.count(fc.when(fc.isnan(c), c)).alias(c) for c in df.columns]).show()
df.agg(*[fc.count(fc.when(fc.isnull(c), c)).alias(c) for c in df.columns]).show()
```

```
df = df.drop('report_location')
```

```
def fill_with_mean(this_df, exclude=set()):
  stats = this_df.agg(*(fc.avg(c).alias(c) for c in this_df.columns if c not in exclude))
  return this_df.na.fill(stats.first().asDict())


df = fill_with_mean(df, ["address", "type", "datetime", "incident_number"])
display(df)
```

```
df.createOrReplaceTempView('Json')
df = spark.sql("SELECT * FROM Json WHERE type in ('Aid Response', 'Medic Response', 'Trans to AMR', 'Auto Fire Alarm')")
```

```
df_final = df.drop('address', 'datetime', 'incident_number')
display(df_final)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 


# Register dataframe as a temp table in SQL context
df_final.createOrReplaceTempView("df")

sqlStatement = """
    SELECT 
    *
    FROM df
"""
plotdata = spark.sql(sqlStatement).toPandas();


# show correlation matrix heatmap to explore some potential interesting patterns
corr = plotdata.corr()
plt.figure(figsize=(15, 15))
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```
features = df_final.drop('type')
features = list(s for s in features.columns)
print(features)
```

##Pipeline

```
indexer = ft.StringIndexer(inputCol="type", outputCol="typeIndex")
indexed = indexer.fit(df_final).transform(df_final)
indexed = indexed.drop('type')
indexed.show()
```

```
train, test = indexed.randomSplit([0.7, 0.3], seed = 1)
print("Distribution of label' value in train is: ", train.groupBy("typeIndex").count().take(3))
print("Distribution of label' value in test is: ", test.groupBy("typeIndex").count().take(3))
```

##Model train and prediction

```
import pyspark.ml.feature as ft
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml import Pipeline

# We choose features
assem = ft.VectorAssembler(inputCols=[col for col in features], outputCol='features')
# Then we build a simple RF classifier with 20 trees
rf = RandomForestClassifier(featuresCol='features', labelCol='typeIndex', numTrees=20, maxDepth=30)
# And put the two steps into a nice pipeline
pipeline = Pipeline(stages=[assem, rf])

# Now we fit the model and use transform to get the predictions
model = pipeline.fit(train)
predictions = model.transform(test)
```

##Evaluetion

```
#!pip install handyspark
from handyspark import *

predictions.toHandy().cols[['probability', 'prediction', 'typeIndex']][:5]
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Let's use the run-of-the-mill evaluator
evaluator = BinaryClassificationEvaluator(labelCol='typeIndex')

# We have only two choices: area under ROC and PR curves :-(
auroc = evaluator.evaluate(predictions, {evaluator.metricName: "areaUnderROC"})
auprc = evaluator.evaluate(predictions, {evaluator.metricName: "areaUnderPR"})
print("Area under ROC Curve: {:.4f}".format(auroc))
print("Area under PR Curve: {:.4f}".format(auprc))
```

```
bcm = BinaryClassificationMetrics(predictions, scoreCol='probability', labelCol='typeIndex')

print("Area under ROC Curve: {:.4f}".format(bcm.areaUnderROC))
print("Area under PR Curve: {:.4f}".format(bcm.areaUnderPR))

# But now we can PLOT both ROC and PR curves!
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
bcm.plot_roc_curve(ax=axs[0])
bcm.plot_pr_curve(ax=axs[1])

# We can also get all metrics (FPR, Recall and Precision) by threshold
bcm.getMetricsByThreshold().filter('fpr between 0.19 and 0.21').toPandas()

# And get the confusion matrix for any threshold we want
bcm.print_confusion_matrix(.415856)
```

```
from sklearn.metrics import classification_report, confusion_matrix

y_true = test.select(['typeIndex']).collect()
y_pred = predictions.select(['prediction']).collect()

print(classification_report(y_true, y_pred))
print(confusion_matrix(y_true, y_pred))
```

#Streaming processing

```
from pyspark.sql.functions import window


# Path to our 20 JSON files
inputPath = "/FileStore/tables/sfd/"

# Explicitly set schema


# Create DataFrame representing data in the JSON files
streamingInputDF = (
  spark
    .readStream
    .schema(schema)               #schema giÃ  definito in the static processing
    .option("maxFilesPerTrigger", 20)  # Treat a sequence of files as a stream by picking one file at a time
    .json(inputPath)
)

streamingCountsDF = (
  streamingInputDF
     .groupBy(
      streamingInputDF.type,
      window(streamingInputDF.datetime, "1 minute"))
    .count()
)
```

```
query = (
  streamingCountsDF
    .writeStream
    .format("memory")        # memory = store in-memory table (for testing only)
    .queryName("jsons")     # counts = name of the in-memory table
    .outputMode("complete")  # complete = all the counts should be in the table
    .start()
)
```
